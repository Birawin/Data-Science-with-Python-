# Analyzing E-Commerce Purchases

```python
#Mean

ecom['Purchase Price'].mean()
ecom['Purchase Price'].mean().astype('float')
ecom['Purchase Price'].mean().astype('int')
```

```python
# How many people have English 'en' as their Language of Choice on the Website?

ecom[(ecom['Language']) == 'en'].count()

#Answer: 1098
```

```python
ecom[(ecom['Language']) == 'en']
```

```python
# How many people have the job title of "Lawyer?"

ecom[(ecom['Job'] == "Lawyer")].info()
```

```python
ecom[(ecom['Job'] == "Lawyer")].head()
```

```python
#How many people made the purchase during the AM and how many people made the purchase during the "PM" hours?

ecom[(ecom['AM or PM'] == 'AM')].info()
```

```python
ecom[(ecom['AM or PM'] == 'PM')].info()
```

```python
#Another Easier Way to Find the Above

ecom['AM or PM'].value_counts()
```

```python
#5 Most Common Job Titles
ecom['Job'].value_counts()
```

```python
#Someone made a purchase that came from Lot: "90 WT", what was the Purchase Price for this transaction?

ecom[(ecom['Lot'] == "90 WT")]['Purchase Price']
```

```python
# What is the email of the person with the following Credit Card Number: 4926532442672853
ecom[ecom['Credit Card']==4926535242672853]['Email']
```

```python
ecom.loc[(ecom['CC Provider'] == 'American Express') & (ecom['Purchase Price'] > 95)]
```

```python
ecom.loc[(ecom['CC Provider'] == 'American Express') & (ecom['Purchase Price'] > 95)].count()
```

```python
#How many people have a credit card that expires in 2025?

ecom[ecom['CC Exp Date'].apply(lambda exp: exp[3:] == '25')]
```

```python
#What are the top 5 most popular email providers/hosts?

#example_email = ecom['Email'].iloc[15]
#example_email.split('@')[1] - splits the email address at the @ sign for the 16th row on the dataset and prints out the 2nd half [as indicated by the '1']

ecom['Email'].apply(lambda email: email.split('@')[1]).value_counts()
```
