```python
#Average (Mean) BasePay of all employees per year? (2011-2014)

salary_data2011 = salary_data[salary_data['Year'] == 2011]
basepay2011 = salary_data2011['BasePay'].mean().astype(int).astype(str)
print("The average BasePay of 2011 is " + basepay2011)

salary_data2012 = salary_data[salary_data['Year'] == 2012]
basepay2012 = salary_data2012['BasePay'].mean().astype(int).astype(str)
print("The average BasePay of 2012 is " + basepay2012)

salary_data2013 = salary_data[salary_data['Year'] == 2013]
basepay2013 = salary_data2013['BasePay'].mean().astype(int).astype(str)
print("The average BasePay of 2013 is " + basepay2013)

salary_data2014 = salary_data[salary_data['Year'] == 2014]
basepay2014 = salary_data2014['BasePay'].mean().astype(int).astype(str)
print("The average BasePay of 2014 is " + basepay2014)
```

```python
#Simplified Version"
salary_data.groupby('Year').mean()['BasePay']
```

```python
#How many People have Chief in their Job Titles

def chief_string(title):
    if 'chief' in title.lower().split():
        return True
    else:
        return False

chief_string("Chief Manager-Metropolitan Transit Authority")
```
