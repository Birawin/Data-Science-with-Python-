K-Means Clustering

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
from sklearn.cluster import KMeans
```

```python
#Import Data

data = pd.read_csv('Countries-exercise.csv')
```

```python
data.head()
```

```python
data.tail()
```

```python
plt.scatter(data['Longitude'], data['Latitude'])
plt.xlim(-180,180)
plt.ylim(-90,90)
plt.show()
```

```python
# Selecting Features

x = data.iloc[:,1:3]
x
```

```python
kmeans = KMeans(7)
kmeans.fit(x)
```

```python
identified_clusters = kmeans.fit_predict(x)
identified_clusters

data_w_cluster = data.copy()
data_w_cluster['Cluster'] = identified_clusters
```

```python
plt.scatter(data_w_cluster['Longitude'], data['Latitude'], c=data_w_cluster['Cluster'], cmap='rainbow')
plt.xlim(-180,180)
plt.ylim(-90,90)
plt.show()
```
